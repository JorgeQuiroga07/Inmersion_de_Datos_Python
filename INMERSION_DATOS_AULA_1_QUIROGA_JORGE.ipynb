```python
Datos_estadistico = inmuebles.groupby(['Barrio'])['Area'].agg(['count', 'median', 'min', 'max'])
Datos_estadistico.head(15)
```

**2.Precio de m2 por barrio y hacer el gráfico más adecuado para esta nueva variable**

```python
 from pickle import NONE
inmuebles_barrio_m2 = inmuebles.groupby('Barrio').mean().head(15)
inmuebles_barrio_m2['Precio_m2'] = inmuebles_barrio_m2['Precio_Millon'] / inmuebles_barrio_m2['Area']
inmuebles_barrio_m2.sort_values(by='Precio_m2', ascending=False, inplace=True)

plt.figure(figsize=(10 ,6))
grafica = sns.barplot(data=inmuebles_barrio_m2, y=inmuebles_barrio_m2.index, x='Precio_m2', palette='coolwarm',orient='h' )

grafica.set_title('DISTRIBUCIÓN DE PRECIO POR METRO CUADRADO', color='blue', fontsize=20)
grafica.set_xlabel('Precio en millones')
grafica.set_ylabel('Barrio')

plt.show()
```
